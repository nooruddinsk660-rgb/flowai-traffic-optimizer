```
from ultralytics import YOLO

model = YOLO('yolov8n.pt')   # Start from pre-trained base

results = model.train(
    data='/content/ambulance_data/data.yaml',
    epochs=60,         # 60 epochs = good accuracy, ~45 min on T4
    imgsz=640,
    batch=16,
    patience=15,       # Early stopping if no improvement for 15 epochs
    save=True,
    name='ambulance_v1',
    exist_ok=True
)
print(f"Best mAP50: {results.results_dict['metrics/mAP50(B)']:.3f}")
```
